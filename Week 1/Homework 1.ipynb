# Question 7

```python
#  Select all the Lotus cars from the dataset
df_lotus = data[data.Make.str.lower() == "lotus"]
df_lotus
```

```python
df_lotus = df_lotus[['Engine HP', 'Engine Cylinders']]
df_lotus
```

```python
df_lotus = df_lotus.drop_duplicates()
df_lotus
```

```python
X = df_lotus.values
X
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inverse = np.linalg.inv(XTX)
XTX_inverse
```

```python
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
y
```

```python
w = (XTX_inverse@(X.T))@y
w
```

The value of w[0] is 4.59494481
